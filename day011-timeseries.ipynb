```python
%matplotlib inline
from hydrophone_data_processing import load, preprocessing
import obspy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
paths = ['/media/sda/data/robdata/Hydrophones/DAYS/A00/A00.7F.04.GDH.2020.011'
         ,'/media/sda/data/robdata/Hydrophones/DAYS/B00/B00.7F.01.GDH.2020.011'
         , '/media/sda/data/robdata/PASSCAL/Main_deployment/DAYS/A01/A01.7F..GHZ.2020.011'
         , '/media/sda/data/robdata/PASSCAL/Main_deployment/DAYS/B01/B01.7F..GHZ.2020.011'
        ]
stream = load.get_raw_stream(paths=paths)
stream[0] = preprocessing.convert_trace_to_pascals(stream[0])
stream[2] = preprocessing.convert_trace_to_pascals(stream[2])
stream.trim(starttime=obspy.UTCDateTime('2020-01-11T12:00:00'))
x = stream.plot()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(15, 15), sharex=True)

data = stream.slice(starttime=obspy.UTCDateTime('2020-01-11T14:00:00')
                    , endtime=obspy.UTCDateTime('2020-01-11T16:00:00'))

ax[0].plot(data[0].times(), data[0].data)
ax[0].text(s='A00.04.GDH', x=0, y=-2, fontsize=15)

ax[1].plot(data[2].times(), data[2].data)
ax[1].text(s='B00.01.GDH', x=0, y=-2, fontsize=15)

ax[2].plot(data[1].times(), data[1].data)
ax[2].text(s='A01..BHZ', x=0, y=-2, fontsize=15)

ax[3].plot(data[3].times(), data[3].data)
ax[3].text(s='B01..BHZ', x=0, y=-2, fontsize=15)

fig.tight_layout()
```
