```python
import obspy
import io
path = '/media/sda/data/borehole/queryauth?net=7F&sta=B07&cha=GH2&starttime=2019-01-01&endtime=2021-01-01&format=miniseed&nodata=404'
reclen = 512
# chunksize = 100000 * reclen # Around 50 MB
chunksize = 10000 * reclen # Around 50 MB
with io.open(path, "rb") as fh:
    while True:
        with io.BytesIO() as buf:
            c = fh.read(chunksize)
            
            if not c:
                break
            buf.write(c)
            buf.seek(0, 0)
            st = obspy.read(buf)
            
        # Do something useful!
        print(st)
        break
```
